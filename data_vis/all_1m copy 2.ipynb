```python
import matplotlib.pyplot as plt
import numpy as np

# Metrics and values
metrics = ['NDCG', 'Hit Ratio', 'Fairness1', 'Fairness2', 'Fairness3', 'Fairness4', 'Fairness5', 'Fairness6']
original_model = [0.1794, 0.9347, 0.0592, 0.0850, 0.0134, 0.0152, 0.0057, 0.0651]
fair_model = [0.1124, 0.8533, 0.0309, 0.0401, 0.0054, 0.0079, 0.0028, 0.0394]

# Combine both models' values for normalization
all_values = np.array(original_model + fair_model)
min_value, max_value = np.min(all_values), np.max(all_values)

# Normalize values for the spider chart
normalized_original = (np.array(original_model) - min_value) / (max_value - min_value)
normalized_fair = (np.array(fair_model) - min_value) / (max_value - min_value)

# Close the loop by appending the first value to the end
normalized_original = np.append(normalized_original, normalized_original[0])
normalized_fair = np.append(normalized_fair, normalized_fair[0])

# Prepare angles for the radar plot
num_metrics = len(metrics)
angles = np.linspace(0, 2 * np.pi, num_metrics, endpoint=False).tolist()
angles += angles[:1]  # Close the loop

# Initialize the radar plot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection': 'polar'})

# Plot the data for both models
ax.plot(angles, normalized_original, label='Original Model', color='blue', linewidth=2)
ax.fill(angles, normalized_original, color='blue', alpha=0.25)

ax.plot(angles, normalized_fair, label='Fairness-Aware Model', color='green', linewidth=2)
ax.fill(angles, normalized_fair, color='green', alpha=0.25)

# Add labels and legend
ax.set_yticks([])  # Hide radial grid labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(metrics, fontsize=10)
ax.set_title('Comparison of Metrics', fontsize=14, pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1), fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```
